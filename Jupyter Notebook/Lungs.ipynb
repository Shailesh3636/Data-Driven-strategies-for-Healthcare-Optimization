```python
df.loc[df['GENDER']=='M','GENDER'] = 1
df.loc[df['GENDER']=='F','GENDER'] = 0
df['GENDER'] = df['GENDER'].astype('int')
df.head()
```

```python
df.loc[df['LUNG_CANCER']=='YES','LUNG_CANCER'] = 1
df.loc[df['LUNG_CANCER']=='NO','LUNG_CANCER'] = 0
df['LUNG_CANCER'] = df['LUNG_CANCER'].astype('int')
df.head()
```

```python
df["SMOKING"].replace({2: 1, 1: 0}, inplace=True)
df["YELLOW_FINGERS"].replace({2: 1, 1: 0}, inplace=True)
df["ANXIETY"].replace({2: 1, 1: 0}, inplace=True)
df["PEER_PRESSURE"].replace({2: 1, 1: 0}, inplace=True)
df["CHRONIC DISEASE"].replace({2: 1, 1: 0}, inplace=True)
df["FATIGUE "].replace({2: 1, 1: 0}, inplace=True)
df["ALLERGY "].replace({2: 1, 1: 0}, inplace=True)
df["WHEEZING"].replace({2: 1, 1: 0}, inplace=True)
df["ALCOHOL CONSUMING"].replace({2: 1, 1: 0}, inplace=True)
df["COUGHING"].replace({2: 1, 1: 0}, inplace=True)
df["SHORTNESS OF BREATH"].replace({2: 1, 1: 0}, inplace=True)
df["SWALLOWING DIFFICULTY"].replace({2: 1, 1: 0}, inplace=True)
df["CHEST PAIN"].replace({2: 1, 1: 0}, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop("LUNG_CANCER", axis = 1)
y = df.LUNG_CANCER

print("Shape of X: ", X.shape)
print("Shape of y: ", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
len(X_train.columns)
```

```python
from sklearn.svm import SVC
svc = SVC(C = 1000, gamma=0.0001)
svc.fit(X_train, y_train)
y_pred2 = svc.predict(X_test)
evaluate_model(y_test, y_pred2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(max_features=0.55, min_samples_leaf=2, random_state=2)
rf_clf.fit(X_train, y_train)
pred5 = rf_clf.predict(X_test)
evaluate_model(y_test, pred5)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbdt = GradientBoostingClassifier(n_estimators = 200, learning_rate = 0.1, max_depth=1, random_state=0)
gbdt.fit(X_train, y_train)
pred_gbdt = gbdt.predict(X_test)
evaluate_model(y_test, pred_gbdt)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
lr_clf_pipe = make_pipeline(StandardScaler(), LogisticRegression())
lr_clf_pipe.fit(X_train, y_train)
y_pred1 = lr_clf_pipe.predict(X_test)
evaluate_model(y_test, y_pred1)
```
